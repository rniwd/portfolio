```python
df = df_combined

def create_scatter_plots(df, target, cols, title):
    fig, axes = plt.subplots(nrows=5, ncols=4, figsize=(20, 15))
    fig.suptitle(title, fontsize=16)
    
    for idx, feature in enumerate(cols):
        ax = axes[idx // 4, idx % 4]
        df.plot(x=feature, y=target, kind="scatter", ax=ax)
        ax.set_title(f"{feature} vs {target}")
        
    for idx in range(len(cols), 20):
        fig.delaxes(axes[idx // 4, idx % 4])
        
    plt.tight_layout()
    plt.subplots_adjust(top=0.95) 
    plt.show()

features = [
    ".header.seq", ".pose.pose.position.x", ".pose.pose.position.y", 
    ".pose.pose.orientation.z", ".pose.pose.orientation.w", 
    ".twist.twist.linear.x", ".twist.twist.angular.z", 
    ".orientation.x", ".orientation.y", ".orientation.z", 
    ".orientation.w", ".angular_velocity.x", ".angular_velocity.y", 
    ".angular_velocity.z", ".linear_acceleration.x", 
    ".linear_acceleration.y", ".linear_acceleration.z"
]

create_scatter_plots(df, ".linear.x", features, "Scatter plots for .linear.x")
create_scatter_plots(df, ".angular.z", features, "Scatter plots for .angular.z")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.tree import plot_tree
class mod1:
    def __init__(self, df_comb):
        self.df1 = df_comb
    def linreg_for_angularz(self):
        X = self.df1[['.angular_velocity.z', '.twist.twist.angular.z']]
        y = self.df1['.angular.z']
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
        lin_reg = LinearRegression()
        lin_reg.fit(X_train, y_train)
        predictions = lin_reg.predict(X_test)
        print(f"accuracy: {lin_reg.score(X_test, y_test)}")

        plt.figure(figsize=(10, 6))
        plt.scatter(y_train, lin_reg.predict(X_train), color='blue', label='Обучающие данные')
        plt.scatter(y_test, predictions, color='red', label='Тестовые данные')
        plt.plot([y.min(), y.max()], [y.min(), y.max()], color='black', linestyle='--', linewidth=2, label='Идеальная модель')            
        plt.xlabel('Фактические')
        plt.ylabel('Предсказанные')
        plt.legend()
        plt.show()
    


    def decisionTree(self):
        y = self.df1[['.linear.x']]
        X = self.df1.drop(['.linear.x'], axis=1)
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
        regr_1 = DecisionTreeRegressor(max_depth=2)
        regr_2 = DecisionTreeRegressor(max_depth=5)
        regr_1.fit(X_train, y_train)
        regr_2.fit(X_train, y_train)
        y_1 = regr_1.predict(X_test)
        y_2 = regr_2.predict(X_test)
        mse = mean_squared_error(y_test, y_1)
        print("MSE:", mse)
        mse = mean_squared_error(y_test, y_2)
        print("MSE2:", mse)

        plt.figure(figsize=(20,10))
        plot_tree(regr_2, filled=True, feature_names=X_train.columns, rounded=True,fontsize=6)
        plt.show()
        plt.figure(figsize=(20,10))
        plot_tree(regr_1, filled=True, feature_names=X_train.columns, rounded=True)
        plt.show()



processor = mod1(df)
processor.linreg_for_angularz()
processor.decisionTree()

```
