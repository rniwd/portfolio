```python
df['Shape__Area'] = df['Shape__Area'].fillna(df['Shape__Area'].mean())
df['Shape__Length'] = df['Shape__Length'].fillna(df['Shape__Length'].mean())
```

```python
import seaborn as snsOBJECTID  NAME          
import matplotlib.pyplot as plt

df_cleaned = df.dropna()
sns.pairplot(df_cleaned, hue="Obesity")
plt.show()
```

```python
df.drop('NAME', axis = 1, inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', square=True, fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
df['Obesity_Rounded'] = df['Obesity'].round(0)
sns.catplot(x='Obesity_Rounded', kind="count", data=df, height=6, aspect=1.5)
plt.title("   ( )")
plt.xticks(rotation=45)
plt.show()
```

```python
bins = [20, 25, 30, 35, 40]
labels = ['20-25', '25-30', '30-35', '35-40']
df['Obesity_Grouped'] = pd.cut(df['Obesity'], bins=bins, labels=labels, include_lowest=True)
sns.catplot(x='Obesity_Grouped', kind="count", data=df, height=6, aspect=1.5)
plt.title("   ( )")
plt.show()
```

```python
df.groupby('NAME')['Obesity'].mean()
```

```python
pivot_table = df.pivot_table(values='Obesity', index='NAME', aggfunc='mean')
pivot_table
```

```python
Q1 = df['Obesity'].quantile(0.25)
Q3 = df['Obesity'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['Obesity'] < (Q1 - 1.5 * IQR)) | (df['Obesity'] > (Q3 + 1.5 * IQR))]
outliers
```
